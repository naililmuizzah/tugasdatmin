# CEK DAN PENANGANAN OUTLIER

**IQR**

```python
import pandas as pd

# Membaca data dari CSV
data = pd.read_csv("data_liver.csv")

# Mengganti nama kolom
data = data.rename(index=str, columns={
    "65": "Ages",
    "Female": "Gender",
    "0.7": "TB",
    "0.1": "DB",
    "187": "AAP",
    "16": "SGPT",
    "18": "SGOT",
    "6.8": "TP",
    "3.3": "ALB",
    "0.9": "A/G",
    "1": "Diagnosis"
})

# Ambil kolom numerik
num_cols = data.select_dtypes(include=['number']).columns

# Fungsi deteksi outlier dengan IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

# Cek jumlah outlier untuk setiap kolom numerik
for col in num_cols:
    outliers = detect_outliers_iqr(data, col)
    print(f"{col}: {len(outliers)} outliers")
```

#Penanganan Outlier

**Winsorizing (Mengganti Outlier dengan Batas IQR)**

```python
def winsorize_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = np.where(data[column] < lower_bound, lower_bound, data[column])
    data[column] = np.where(data[column] > upper_bound, upper_bound, data[column])

# Terapkan Winsorizing
for col in num_cols:
    winsorize_iqr(data, col)
```

#Proses membuat model dengan Neural Network

```python
#Menguji parameter yang paling optimum untuk model

from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV

from sklearn.model_selection import StratifiedKFold

cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=100)



# Tentukan parameter yang ingin diuji
parameters = {
    'hidden_layer_sizes': [(50,), (100,), (50, 30), (100, 50)],
    'learning_rate_init': [0.001, 0.01, 0.1],
    'max_iter': [100, 200, 300]
}

# Buat model MLPClassifier
clf = MLPClassifier(random_state=100)

# Inisialisasi GridSearchCV
grid_search = GridSearchCV(clf, parameters, cv=cv_strategy)
# cv merupakan jumlah lipatan validasi silang (cross-validation)

# Lakukan pencarian grid pada data training
grid_search.fit(X, y)
```

```python
# Tampilkan parameter terbaik
print("Parameter terbaik:", grid_search.best_params_)
```

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(hidden_layer_sizes=(50,),learning_rate_init=0.1,max_iter=100)

clf.fit(X_train, y_train)
```

Kode yang disediakan menggunakan MLPClassifier dari library Scikit-Learn untuk membuat model Neural Network berbasis Multi-Layer Perceptron (MLP) dengan parameter tertentu dan melatihnya pada data training.

Dalam kasus ini, model MLPClassifier dibuat dengan parameter-parameter tertentu yang telah ditentukan secara manual. Parameter-parameter ini adalah:

hidden_layer_sizes=(50,): Ini adalah tuple yang menunjukkan arsitektur dari jaringan saraf. (50,) berarti terdapat satu hidden layersberisi 50 neuron.

learning_rate_init=0.1: Parameter ini adalah tingkat pembelajaran awal. Ini menunjukkan seberapa besar model akan menyesuaikan bobotnya berdasarkan kesalahan pada setiap iterasi.

max_iter=100: Parameter ini menunjukkan jumlah maksimum iterasi yang akan dilakukan saat melatih model. Ini mengontrol seberapa banyak iterasi atau epoch yang akan dilakukan selama proses pelatihan.

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

Hasil dari laporan klasifikasi tersebut memberikan ringkasan performa model Neural Network pada data uji (liver) dengan menggunakan berbagai metrik evaluasi seperti precision, recall, dan f1-score. Berikut adalah interpretasi dari setiap metrik yang tercantum dalam laporan:

Precision: 1 (Kelas Negatif): Dari semua prediksi yang dilakukan sebagai kelas negatif, sekitar 77% di antaranya adalah benar-benar kelas negatif. 2 (Kelas Positif): Sekitar 57% dari semua prediksi yang dilakukan sebagai kelas positif adalah benar.

Recall: 1 (Kelas Negatif): Dari semua sampel yang sebenarnya adalah kelas negatif, model mampu mengenali sekitar 93% dari mereka. 2 (Kelas Positif): Dari semua sampel yang sebenarnya adalah kelas positif, model hanya dapat mengenali sekitar 25% dari mereka.

F1-Score: F1-score adalah ukuran rata-rata harmonis dari precision dan recall. F1-score untuk kelas 1 adalah 0.84, sedangkan untuk kelas 2 adalah 0.35. F1-score semakin tinggi semakin baik, karena mencoba untuk menggabungkan keseimbangan antara precision dan recall.

Accuracy: Akurasi model adalah sekitar 74%, yang mengindikasikan seberapa baik model dapat memprediksi dengan benar secara keseluruhan dari seluruh sampel.

Interpretasi: Meskipun akurasi secara keseluruhan adalah 74%, melihat pada recall untuk kelas 2 (0.25) menunjukkan bahwa model memiliki performa yang lebih rendah dalam mengidentifikasi kasus positif (2). Terlihat bahwa model memiliki kemampuan yang baik dalam memprediksi kasus negatif (1) tetapi kurang baik dalam memprediksi kasus positif (2). Interpretasi ini penting tergantung pada konteks aplikasi, karena keseimbangan antara kelas 1 dan 2 bisa menjadi kunci dalam pengambilan keputusan yang dihasilkan oleh model.
